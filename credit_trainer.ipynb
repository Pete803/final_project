```python
data_df = all_applicants.drop(['job_title', 'sex'],  axis=1)
data_df.head()
```

```python
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
```

```python
data_df['salary'] = data_df['salary'].apply(clean_currency).astype('float')
data_df['loan_amount'] = data_df['loan_amount'].apply(clean_currency).astype('float')
data_df['existing_debt'] = data_df['existing_debt'].apply(clean_currency).astype('float')
data_df
```

```python
data_df.dtypes
```

```python
data_df["DTI"] = (
    data_df["existing_debt"]/data_df["salary"])
```

```python
decimals = 2
data_df['DTI'] = data_df['DTI'].apply(lambda x: round(x, decimals))
data_df.head()
```

```python
slim_data = data_df.drop(['first_name', 'last_name', 'state'], axis=1)
slim_data
```

```python
 # Create the bins in which Data will be held
# Bins are 0, 619, 620, 850   
score_bins = [0, 620, 850]
preapproval = ["Denied", "PreApproved"]
```

```python
slim_data["PreApprovals"] = pd.cut(slim_data["credit_score"], score_bins, labels=preapproval, include_lowest=True)
slim_data
```

```python
first_cut = slim_data.loc[slim_data["PreApprovals"] == "PreApproved", :]
first_cut.head()
```

```python
first_cut.to_csv('Resources/first_cut.csv', index=False, header=True)
```

```python
second_cut = first_cut.drop('PreApprovals',  axis=1)
second_cut.head()
```

```python
#slim_data[slim_data['DTI']==slim_data['DTI'].max()]
```

```python
dti_bins = [0, 0.36, 2.76]
finalapproval = ["Approved", "Denied"]
```

```python
second_cut["Approvals"] = pd.cut(second_cut["DTI"], dti_bins, labels=finalapproval, include_lowest=True)
second_cut
```

```python
second_cut.to_csv('Resources/second_cut.csv', index=False, header=True)
```

```python
import tensorflow
tensorflow.keras.__version__
```

```python
financials = second_cut
financials
```

```python
X = financials.drop("Approvals", axis=1)
y = financials["Approvals"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        X, y, random_state=1)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
# Create the deep learning model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Create model and add Layers
model = Sequential()
model.add(Dense(units=80, activation='relu', input_dim=9))
model.add(Dense(units=80, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
# Quantify our trained model
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Make Predictions
encoded_predictions = model.predict_classes(X_test_scaled[:27])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:27])}")
```
