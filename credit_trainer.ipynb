```python
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
```

```python
data_df['salary'] = data_df['salary'].apply(clean_currency).astype('float')
data_df['loan_amount'] = data_df['loan_amount'].apply(clean_currency).astype('float')
data_df['existing_debt'] = data_df['existing_debt'].apply(clean_currency).astype('float')
data_df
```

```python
data_df.dtypes
```

```python
data_df["DTI"] = (
    data_df["existing_debt"]/data_df["salary"])
```
